```python
import folium
import pandas as pd
from folium.plugins import MarkerCluster
from folium.plugins import MousePosition
from folium.features import DivIcon
```

```python
spacex_df=pd.read_csv('spacex_launch_geo.csv')
spacex_df
```

```python
spacex_geo_data = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]
launch_sites_df
```

```python
#Initializing folium map with default point to displaya as NASA.
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)
site_map
```

```python
# Creating a blue circle at NASA Johnson Space Center's coordinate with a popup label showing its name
circle = folium.Circle(nasa_coordinate, radius=1000, color='blue', fill=True).add_child(folium.Popup('NASA Johnson Space Center'))
# Creating a blue circle at NASA Johnson Space Center's coordinate with a icon showing its name
marker = folium.map.Marker(
    nasa_coordinate,
    # Create an icon as a text label
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'NASA JSC',
        )
    )
site_map.add_child(circle)
site_map.add_child(marker)
```

```python
# showing thelaunch sites in the map.
coordinates = {'ccafs_lc':[28.562302,-80.577356],'ccafs_slc':[28.563197,-80.576820],'ksc_lc':[28.573255,-80.646895],'vafb_slc':[34.632834,-120.610745]}
```

```python
for location, coords in coordinates.items():
    latitude, longitude = coords
    
    # Create a circle at the coordinates
    circle = folium.Circle(location=[latitude, longitude], radius=1000, color='orange', fill=True)
    circle.add_child(folium.Popup(location))
    
    # Create a marker with a custom label
    marker = folium.Marker(
        location=[latitude, longitude],
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % location,
        )
    )
    
    # Add the circle and marker to the map
    site_map.add_child(circle)
    site_map.add_child(marker)
site_map
```

I now try to enhance the map by adding the launch outcomes for each site, and seeing which sites have high success rates. The data frame spacex_df has detailed launch records, and the class column indicates if this launch was successful or not.

```python
#Class=1, success else failure.
spacex_df.tail(5)
```

I will first create markers for all launch records. If a launch was successful (class=1), then I use a green marker and if a launch was failed, I use a red marker (class=0).
Since, a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.

```python
#Creating marker cluster object.
marker_cluster = MarkerCluster()
```

```python
for index, row in spacex_df.iterrows():
    site_name = row['Launch Site']
    latitude = row['Lat']
    longitude = row['Long']
    marker_color = 'green' if row['class'] == 1 else 'red'
    
    # Creating a marker and adding it to the cluster
    marker = folium.Marker(
        location=[latitude, longitude],
        icon=folium.Icon(color=marker_color)
    )
    marker_cluster.add_child(marker)

# Adding the MarkerCluster to the map
site_map.add_child(marker_cluster)
```

The above map can be interacted with by clicking the circles until they zoom and show the successful and failed attempts of each site.
